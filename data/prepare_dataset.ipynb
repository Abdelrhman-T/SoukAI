```python
import re

def preprocess_review(text):
    if not isinstance(text, str):
        return ""

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', ' ', text)

    # Remove Emails
    text = re.sub(r'\S+@\S+', ' ', text)

    # Remove Arabic Diacritics
    arabic_diacritics = re.compile("""
        ّ|َ|ً|ُ|ٌ|ِ|ٍ|ْ|ـ
    """, re.VERBOSE)
    text = re.sub(arabic_diacritics, '', text)

    # Normalize Arabic letters
    text = re.sub(r'[إأآا]', 'ا', text)
    text = re.sub(r'ى', 'ي', text)
    text = re.sub(r'ة', 'ه', text)

    # Remove English characters
    text = re.sub(r'[A-Za-z]', ' ', text)

    # Keep Arabic letters, digits, spaces
    text = re.sub(r'[^\u0600-\u06FF0-9\s]', ' ', text)

    # Remove repeated characters (جمييييل -> جميل)
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    return text
```

```python
def compare(dataset, coulmn, index):
  print(f"Before: {dataset[coulmn][index]}")
  print(f"After: {preprocess_review(dataset[coulmn][index])}")
```

```python
compare(test_dataset, 'review', 0)
```

```python
compare(test_dataset, 'review', 45)
```

```python
def csv_to_json(df):
    records = []
    for _, row in df.iterrows():
        records.append({
            "review": row["review"],
            "intent": row["intent"],
            "urgency": row["urgency"],"sentiment": row["sentiment"],
            "dialect": row["dialect"],
            "routed_team": row["routed_team"],"requires_human": row["requires_human"],
            "entities": {
                "order_id": row["order_id"],"product": row["product"],
                "amount": row["amount"],"date": row["date"]
            }
        })

    return records
```

```python
Json_test_dataset = csv_to_json(test_dataset)
```
